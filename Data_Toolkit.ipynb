2.How does broadcasting work in NumPy?
  - Broadcasting in NumPy is a powerful mechanism that allows arithmetic operations to be performed on arrays of different shapes and sizes without explicitly reshaping them. When two arrays are involved in an operation, NumPy compares their shapes element-wise, starting from the trailing dimensions. If the dimensions are equal or one of them is 1, they are considered compatible. NumPy then "broadcasts" the smaller array across the larger one by virtually replicating its data along the missing dimensions, so their shapes match. This avoids unnecessary memory usage and enables efficient element-wise operations. Broadcasting simplifies code and improves performance, especially when working with large datasets, as it eliminates the need for explicit loops or array replication. It is widely used in data manipulation, machine learning, and scientific computations.

6.What are the differences between NumPy arrays and Python lists?
  - NumPy arrays and Python lists are both used to store collections of items, but they differ significantly in performance, functionality, and use cases. A NumPy array is a homogeneous, multi-dimensional container of items of the same data type, designed specifically for numerical computation, whereas a Python list is a more general-purpose, heterogeneous container that can store items of different types. NumPy arrays provide fast and efficient storage and operations because they are implemented in C, allowing element-wise operations and advanced mathematical functions to be performed on entire arrays without the need for explicit loops. In contrast, Python lists are slower for numerical computations because operations on them are interpreted in Python and require explicit loops. NumPy also provides powerful functionalities like broadcasting, vectorized operations, and convenient methods for linear algebra and statistics, which are not available for Python lists. Overall, NumPy arrays are preferred when working with large-scale numerical data, while Python lists are more suitable for general-purpose programming tasks.

8.What does the term “vectorized operation” mean in NumPy?
  - The term “vectorized operation” in NumPy refers to performing element-wise operations on entire arrays without using explicit Python loops. Instead of iterating over each element individually, NumPy applies operations directly to whole arrays in a highly efficient manner using optimized, low-level C implementations. This allows mathematical operations such as addition, subtraction, multiplication, or more complex functions to be applied to every element of an array in a single step, improving both code readability and performance. For example, adding two NumPy arrays of the same shape simply requires writing array1 + array2, and the operation is applied element-wise automatically. Vectorized operations are faster than using explicit for-loops because they minimize overhead, leverage CPU optimizations, and make use of contiguous memory storage in arrays. This makes vectorization a core concept in scientific computing and data analysis, especially when dealing with large datasets or performing intensive numerical computations.

17.Explain the difference between apply() and map() in Pandas.
   - In Pandas, both apply() and map() are used to transform data, but they differ in their scope and flexibility. The map() function is primarily used with a Pandas Series to apply a function element-wise, replacing each value with the result of the function, dictionary mapping, or a Series. It is best suited for simple, element-wise transformations. On the other hand, apply() is more versatile and can be used on both Series and DataFrames. When used on a Series, it works similarly to map(), but when applied to a DataFrame, it can operate along either rows or columns, allowing aggregation, complex transformations, or custom functions that consider multiple columns at once. In essence, map() is simpler and faster for straightforward, element-wise operations, while apply() provides greater flexibility for more complex and multi-dimensional transformations.

21.Why is NumPy’s array slicing faster than Python’s list slicing?
   - NumPy’s array slicing is faster than Python’s list slicing because NumPy arrays are stored in contiguous blocks of memory and are implemented in low-level C code, allowing operations to be executed at a much lower computational cost. When you slice a NumPy array, it returns a view of the original data rather than creating a new copy, which avoids unnecessary memory allocation and data copying. In contrast, Python lists are collections of object references stored non-contiguously, so slicing a list involves creating a new list and copying each element individually, which is slower. Additionally, NumPy leverages vectorized operations and optimized memory access patterns, enabling efficient computation over large datasets. This combination of contiguous storage, low-level implementation, and avoidance of redundant copying makes NumPy slicing highly performant compared to standard Python list slicing.

```python
#Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [5, 3, 6, 2, 7]
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Q11.Perform matrix multiplication using NumPy.
import numpy as np

# Create two 2D arrays (matrices)
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication
C = np.dot(A, B)
# Alternatively, you can use: C = A @ B

print("Matrix A:")
print(A)
print("Matrix B:")
print(B)
print("Matrix Multiplication Result (A x B):")
print(C)
```

```python
#Q13. Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 15, 13, 17, 20],
    'Z': [5, 8, 12, 6, 9],
    'Category': ['A', 'B', 'A', 'B', 'A']
}

df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = px.scatter_3d(
    df,
    x='X',
    y='Y',
    z='Z',
    color='Category',  # Color points by category
    size='Y',          # Size points based on Y values
    symbol='Category', # Different symbols for categories
    title='3D Scatter Plot Example'
)

# Show the plot
fig.show()
```
